```python
salaries = [
    ("Mark", 1000),
    ("John", 1500),
    ("Daniel", 2300),
    ("Greg", 5000)
]

# %matplotlib inline
# %plotly inline
import matplotlib.pyplot as plt
# import chart_studio.plotly as py
import plotly.graph_objects as go
```

```python
X = [0,1,2,3]
Y = [0,1,0,1]
plt.plot(X, Y)
```

```python
X = [0,1,2,3,4,5,6,7]
Y = [0,1,1,0,0,1,1,0]
plt.plot(X, Y)
```

```python
names = list(map(lambda tup:tup[0], salaries))
salary_values = list(map(lambda tup:tup[1], salaries))
# print(names)
# print(salary_values)
# plt.plot(salary_values)
plt.title("Salaries")
plt.xlabel("Employee")
plt.ylabel("Salary in zł")
plt.bar(names, salary_values)
```

```python
figure, ax = plt.subplots()
ax.bar(names, salary_values)
ax.set_xlabel("Employee")
ax.set_ylabel("Salary in zł")
ax.set_title("Salaries")
```

```python
figure, ax = plt.subplots()
ax.barh(names, salary_values, color=['green'])
ax.set_xlabel("Employee")
ax.set_ylabel("Salary in zł")
ax.set_title("Salaries")
```

```python
X = [0,1,2,3]
Y = [0,1,0,1]

data = go.Scatter(x=X, y=Y)
layout = {
	'title': 'Zigzag with plotly'
}
fig = go.Figure(data, layout)
fig.show()
```

```python
salaries = [
    ("Mark", 1000),
    ("John", 1500),
    ("Daniel", 2300),
    ("Greg", 5000)
]
names = list(map(lambda tup:tup[0], salaries))
salary_values = list(map(lambda tup:tup[1], salaries))

data = [go.Bar(x=names, y=salary_values)]
layout = {
	'title': 'Salaries with plotly'
}
fig = go.Figure(data, layout)
fig.show()
```
